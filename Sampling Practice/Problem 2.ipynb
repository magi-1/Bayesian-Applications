# Bootstrap Filter Example


*Validating my implimentation of the basic Boostrap filter*

Suppose we have 5 independent observations (8, 11, 10, 14, 5) from
a $\text{Poisson}(\theta)$ distribution. Consider a $\text{Gamma}(10,1)$ prior on $\theta$, whence the posterior distribution is $\text{Gamma}(58,6)$ with posterior mean 9.67 and posterior standard deviation 1.27.

```python
class Bootstrap:
  def __init__(self, density, samples, prior_samples, importance = False):
    self.density = density 
    self.samples = samples
    self.prior_samples = prior_samples
    self.importance = importance
    self.weights = self.make_weights()

  def make_weights(self):
    # initializing likelihood
    likelihood = lambda param: np.prod([self.density(x, param) for x in self.samples])
    if self.importance == False: # basic bootstrap filter
      partial_weights = [likelihood(param) for param in self.prior_samples]
      norm = np.sum([partial_weights])
    else: # boostrap filter with importance sampling
      pi, q = self.importance # (prior, importance) densities
      partial_weights = [likelihood(param)*pi(param)/q(param) for param in self.prior_samples]
      norm = np.sum([partial_weights])
    return partial_weights/norm

  def posterior(self):
    # bootstrap estimate of posterior
    estimate = list(zip(self.prior_samples, self.weights))
    df = pd.DataFrame({i: estimate[i] for i in range(len(estimate))}).T
    df.sort_values(by = list(range(0, df.shape[1]-1)), inplace = True)
    return list(zip(*df.values))

  def get_statistics(self):
    # first and second moments
    mean = np.dot(self.weights, self.prior_samples)
    std = np.dot(self.weights, (self.prior_samples-mean)**2)
    return mean, std

  def sample(self, N):
    # sampling from discrete distribution
    support, density = self.posterior()
    return  np.random.choice(support, p = density, size = N)
```

```python
# Poisson observations and prior samples
samples = np.array([8,11,10,14,5])
prior_samples = np.random.gamma(10, 1, 10**4)

B = Bootstrap(poisson.pmf, samples, prior_samples)
support, pmf = B.posterior()
mean, std = B.get_statistics()
```

```python
plt.figure(figsize = (6,5))
plt.plot(support, pmf, color = 'k', linewidth = 3)
plt.title(f'Mean = {round(mean, 4)} | Stdev = {round(np.sqrt(std), 4)}')
plt.show()

print('\nTrue Mean / Std = 9.67, 1.27')
```

# Problem 1

I have observed 20 independent samples from a $\mathcal{N}( |\theta|, 1)$ distribution with sample mean 1.65. Assume a $\mathcal{N}(0, 5^2)$ prior on $\theta$. Run a Bootstrap filter with $T = 10^5$ samples from the prior distribution. Obtain 5000 independent samples by resampling from the obtained discrete approximation to the posterior. Use these samples to draw a histogram of the (approximated) posterior distribution of $\theta$.

```python
class Gaussian:
  # Gaussian with abs(mu) parameterization
  def pdf(x, param, std = 1):
    mu = np.abs(param)
    return np.exp(-(x-mu)**2/(2*std**2))/np.sqrt(2*np.pi*std**2)

prior_samples = np.random.normal(0, 5**2, 10**5)
samples = np.random.normal(1.65, 1, 20)

B = Bootstrap(Gaussian.pdf, samples, prior_samples)
support, pmf = B.posterior()
```

```python
# generating samples
samples = B.sample(N = 5000)

plt.figure(figsize = (6,4))
plt.plot(support, pmf, color = 'k', linewidth = 3)
plt.title('Discrete Estimate of Posterior for Theta')
plt.xlim([-6,6])
plt.show()

plt.figure(figsize = (6.5,4))
plt.hist(samples, bins = 30, color = 'k', density = True)
plt.title('Samples')
plt.xlim([-6,6])
plt.show()
```

# Problem 2

Model a $\text{Beta}(2, 1)$ distribution as a mixture of independent Betas with a uniform prior on $v$

$$\text{Beta}(2, 1) \; \approx \; (1-v)\,\text{Beta}(\frac{3}{2}, 1) + v \,\text{Beta}(\frac{5}{2}, 1)$$ 
$$ v \sim \text{Uniform}(0,1)$$

```python
class BetaMixture:
  def pdf(x, v):
    return (1-v)*beta.pdf(x, 3/2, 1) + v*beta.pdf(x, 5/2, 1)

prior_samples = np.random.uniform(0, 1, 10**5)
samples = np.random.beta(2, 1, 400)

B = Bootstrap(BetaMixture.pdf, samples, prior_samples)
support, pmf = B.posterior()
mean, std = B.get_statistics()
```



I sample data from the target distribution, construct the density function for the mixture of interest, and perform inference on $v$.

```python
# generating samples
samples = B.sample(N = 5000)

plt.figure(figsize = (6,4))
plt.plot(support, pmf, color = 'k', linewidth = 3)
plt.title('Discrete Estimate of Posterior for v')
plt.show()

plt.figure(figsize = (6.5,4))
plt.hist(samples, bins = 50, color = 'k', density = True)
plt.title(f'Mean = {round(mean,4)} | Stdev = {round(std,4)}')
plt.xlim([0,1])
plt.show()
```

See that each component Beta distribibtion either over/under assigns probability mass with respect to $\text{Beta}(2,1)$ by about the same amount in their respective high mass mass subsets of [0,1]. Thus, the estimated posterior mean $v_{\mu} = X$ makes intuitive sense because it can be interpreted as sharing equal amounts of information between the two component Beta distributions.

Consider the data from the Longnecker et al. (2001, Lancet) study relating maternal serum concentration of the DDE metabolite DDE to the risk of preterm birth. Consider fitting a Bayesian logistic regression

  1. Bivariate normal importance density  density $\mathcal{N}(\mu, \Sigma)$ obtained from Berstein-von Mises. 
  2. A bivariate-t importance density $t_v(\mu, \Sigma)$ with $v = 3$ and $\mu$ and $\Sigma$ same as in case 1.

Loading in the data

```python
data = pd.read_csv('DDE.txt', delimiter=' ')
data['intercept'] = np.ones(len(data))

y = data['y'].values
X = data[['intercept', 'x']].values
data.head()
```

Now we need to get the bivariate normal importance density $\mathcal{N}(\mu, \Sigma)$ where $\mu$ is the MLE and $\Sigma$ is the fisher information. Note that I decided to compute the fisher information numerically using JAX's automatic differentiation functionality. I validated my code by computing the closed form solution.

```python
from functools import partial
import jax.numpy as jnp
from jax import grad, jacfwd, jacrev, vmap, jit
from jax.nn import sigmoid

class Bernoulli:
  def pdf(x_i, b):
    return sigmoid(jnp.dot(x_i, b))

  def likelihood(b, X):
    return expit(np.dot(X, b))

class BernsteinVM:
  def __init__(self, X, y):
    self.X = X
    self.y = y

  def MLE(self):
    model = LogisticRegression(random_state=7, penalty = 'none').fit(self.X, self.y)
    return model.coef_[0]
    
  def log_likelihood(self, b, X, y):
    return jnp.sum(jnp.dot(y, jnp.dot(X, b)) - jnp.log(1+jnp.exp(jnp.dot(X, b))))

  def fisher_information(self, mle):
    L = partial(self.log_likelihood, X = self.X, y = self.y)
    return -jacfwd(jacrev(L))(mle)

  def importance_density(self):
    mu = self.MLE()
    cov = self.fisher_information(mu)
    return mu, cov/self.X.shape[0], multivariate_normal(mu, cov).pdf
```

```python
BVM = BernsteinVM(X,y)
mu, cov, q = BVM.importance_density()
pi = multivariate_normal([0,0], 9*np.eye(2)).pdf
prior_samples = np.random.multivariate_normal(mu, cov, 10**5)
```

```python
B = Bootstrap(Bernoulli.pdf, X, prior_samples, (pi, q))
support, pmf = B.posterior()
```
